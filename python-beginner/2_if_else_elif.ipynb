## Python里的是与非
- 是 / 非的值就是 True / False
- 可以通过算子比较来得出是非值

````python
a = True
b = False
print(a)
print(b)
````

### 是非比较算子
```
- x == y    x 等于 y
- x != y    x 不等于 y
- x > y     x 大于 y
- x >= y    x 大于等于 y
- x < y     x 小于 y
- x <= y    x 小于等于 y
```

````python
x = 10
y = x > 5
print(x)
print(y)
````

````python
x = 'hello'
print(x == 'hello')
````

````python
x = 1
y = 'world'
print(x == y)
print(x != y)
````

### 是非逻辑算子
```
- and   和
- or    或
- not   非
```

````python
x = 10
print(x > 1 and x > 10)
print(x > 1 or x > 10)
print(not x > 100)          # x > 100 == False => not (x > 100) == True
x = 100
print(x > 10 and x < 1000 and x / 20 >= 5)  # 逻辑算子能够联合多个条件
````

### 是非对于其他数据类型的换变

````python
# 数字类的是非换变，为0的数字为非
print(bool(0))
print(bool(100))
print(bool(2.32))

# 文字的是非换变，长度为零的字符串为非
print(bool(''))
print(bool('hello'))

# 列的是非换变, 没有元素为非
print(bool([]))
print(bool(['hello', 'world']))
print(bool({}))
````

## if-else-elif
if, else, elif 在程序的作用就是依据条件的是非值来决定是否运行一部分编程
```
- if    如果
- else  或者
- elif  或者如果
```

````python
age = 14
have_license = True
if age > 16 and have_license:    # 利用 : 来表示条件结束
    print('可以驾车了')
else:
    if have_license:
        print('你是怎样拿到license的？') 
    print('不可以驾车!')
````

> 行的缩进级别决定是否在 if else elif 的条件决定会不会运行   
如果 if else elif 以下的编程的缩进级别小于自身的缩进级别，条件成立后就会运行以下的编程
```py
age = 14                                  # 缩进级别: 0
have_license = True
if age > 16 and have_license: {
    print('可以驾车了')                     # 缩进级别: 1
} else: {
    if have_license: {
        print('你是怎样拿到license的？')     # 缩进级别: 2
        print('很神奇')                    # 缩进级别: 2
    }
    print('不可以驾车!')                    # 缩进级别: 1
}

```

````python
age = 123
if age == 21:
    print('你早一年出生就可以拿到tng 150块')
elif age > 21:                      
    print('你拿不到 tng 150')
else: 
    print('你可以拿到 tng 150')
````

````python
money = 10
apple_price = 5
have_apple = False

if money - apple_price > 0:
    print('我买了苹果, yay!')
    money -= apple_price
    have_apple = True

if have_apple:
    print('我有apple了, yay!')
else:
    print('我没有apple, sad')

if money != 0:
    print(f'我剩下{money}块钱.')
else:
    print('我没钱了！')

````
