# CH: 4 For 和 While 循环
在介绍循环的概念前，先来探讨一个问题
> 今天想用程序对一序列人打招呼，在没有循环的概念下，要如何编程这个程序呢？

````python
people = ['john', 'steve', 'jacky', 'mansa', 'aisha']
#  要对每个人打招呼
print(f'hello {people[0]}!')
print(f'hello {people[1]}!')
print(f'hello {people[2]}!')
print(f'hello {people[3]}!')
print(f'hello {people[4]}!')
````

今天如果`people`包含了100个不同个人, 不可能一个一个慢慢写出来  
因此在这样的情况下，我们就会运用for-循环

````python
people = ['john', 'steve', 'jacky', 'mansa', 'aisha']
# for person in people: 意思就是, 
# 对于people中的元素进行循环，每次把被循环的元素值放入名为person的变量

for person in people:           # 也像if else一样，用 : 来终结 for - loop 的定义 
    print(f'hello {person}!') 
````

````python
# for - loop 的对象也要适合进行循环
x = 1
for idk in x:
    print(idk)
````

循环过程的控制
```
- continue  直接进入下次循环
- break     跳出循环
```

````python
favorite_fruit = 'apple'
fruits = ['banana', 'orange', 'tomato', 'peach', 'grapes', 'apple', 'kiwi']
have_favorite_fruit = False

for fruit in fruits:
    print(fruit)
    if fruit == 'tomato':
        print('连水果都不是，直接skip')
        continue                        # 跳到下次循环
    if fruit == favorite_fruit:
        print('yay，有我最爱的水果')
        have_favorite_fruit = True
        break                           # 直接跳出循环
    print('不是我最爱的水果') 

if have_favorite_fruit:
    print('yay！')
````

````python
# range(100) 实际上就是一个返回 [0, 1, 2, ..., 99] 的函数
for i in range(100):
    print(i + 1)
````

### for - 循环中有 for - 循环

````python
people = [
    {
        'name': 'jacky',
        'debts': [100, 200]
    },
    {
        'name': 'john',
        'debts': [100, 200, 300, 500]
    }
]

total_debt = 0
for person in people:
    name = person['name']
    current_person_debt = 0
    
    for debt in person['debts']:
        current_person_debt += debt

    print(f'{name}欠{current_person_debt}块钱，快还钱！')
    total_debt += current_person_debt

print(f'{len(people)}个人欠{total_debt}, 快还钱！')
````

## While 循环
特点：
- 是依据一个条件决定是否继续循环，而不是一个序列的长度
- 通常用于在不能提前知道终结点的过程

````python
# 利用要计算出一个数字要除2几次才能小于1作为例子
# while number > 1: 的意思是：
# 当 number 大于 1 时，持续循环运行 

number = 10
divide_count = 0

while number > 1:       # 也是用 : 来终结
    divide_count += 1
    number /= 2

print(divide_count)
````

while 循环控制
- 也像 for 循环一样, 有 continue，break
```
- continue  直接进入下次循环
- break     跳出循环
```

````python
squared_number = 225

num = 0
while True:
    num += 1
    if num % 2 == 0:        # 是双数可以直接跳过，因为肯定不是它的方根
        continue            
    if (num * num) == squared_number:
        print(f'方根是 {num}')
        break               # 找到了方根，跳出循环
````
