# CH 1: 变量 Variables

特点：
- 变量是储存数据的最小单位
- 变量储存的数据是有分类型的
- 变量所储存的数据是可以改变的

````python
# 这边就是定义变量的方式
# a = 10 的意思是:
# 我们设立了一个变量，名为 a， 它的值是 10

a = 10

# 我们可以直接通过利用变量的名字来获取变量的值
print(a)
````

## 变量的数据种类
主要的数据种类有:
```
- int   整数 e.g. 100, 1, 123
- float 实数 e.g. 10.2, 3.79
- str   文字/字符串 e.g. 'hello', 'mark', 'string', '100'
- bool  是 / 非 e.g. True, False
- list  序列 e.g. [1, 2, 3], ['hello', 'world']
- dict  字典 e.g. { 'john': 100, 'mark': 50 }
``` 

````python
a = 10
b = 3.79
c = 'hello world'
d = False
e = ['hello', 10, True]
f = { 'john': 100, 'mark': 50 }
````

## 变量数据的改变
主要分成两个方式
- 重新定值
- 算子运用 ( 加减乘除等 )

````python
# 重新定值
x = 100
print(x)
x = x * 2
print(x)
x = 'hello' 
print(x)
# x 从 100 变去了 'hello'
````

### 一般的算子类型
```
- +  加
- -  减
- *  乘
- /  除（有小数）
- // 整除（无小数）
- %  取剩值 (mod)
```

````python
x = 20
print(x)
x += 30
print(x)
x -= 50
print(x)
x = 10
x *= 10
print(x)
````

### 变量数据类型的变换

````python
x = int('10')   # int('10') 就是把文字数据换变去整数为10的值
x += 10 
print(x)
x = int(1.232)
print(x)        # 在换变实数去整数时也会去下整数值
x = str(x)      # 换变去 '1'
x += '11'
print(x)
````

````python
# 在换变数据类型时也要确保数据时适合被换变的
x = int('hello')
````

## 文字 / 字符串
这边会简单介绍文字的格式化

````python
money = 100000
age = 21
print('我今年' + str(age) + '岁，我有' + str(money) + '块钱.')
````

````python
# 这个叫 f-字符串
# 定义 f-字符串 的方式 就是在一般文字数据前头添加 f
# {} 里面的数据值会自动被格式化，纳入文字中
print(f'我今年{age}岁，我有{money}块钱.')
print(f'五年后我是{age + 5}岁.')
````
