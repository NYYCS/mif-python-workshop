## 序列
- 能够存储一序列数据，且不分数据类型
- 是有序的，即数据会依据纳入序列的次序进行排列

````python
a = [1, 2, 3, 4, 5, 6]
````

序列所储存的数据不强制数据类型

````python
# 这些都是合法的序列
x = [1, '222', 3]
y = ['hello', 12.123, False]
z = [1, [2, 3, 4], 'world']
````

序列是有序的

````python
x = [1, 2, 3]
y = [3, 2, 1]
print(x == y)   # 非同序，因此不相等
z = [1, 2, 3]
print(x == z)   # 同序，因此相等
````

### 获取序列中的元素
- 获取序列中在指定位置的元素
>序列的位置是从0开始算起的，所以第一个元素其实是以0作为位置的表示

````python
x = [4, 5, 6]
print(x[0])     # x[0] 的意思就是获取x序列第(0 + 1)位置的元素
print(x[2])     # x[2] 的意思就是获取x序列第(2 + 1)位置的元素
print(x[0 + 1]) # 方挂里也能进行加减乘除
````

````python
print(x[3])    # 获取元素是位置不能超于序列的长度 - 1
````

````python
# 负数的位置就是从后面开始算起位置的元素
# x[-1] => x[len(x) - 1] => x[2]
# x[-2] => x[len(x) - 2] => x[1]
# x[-3] => x[len(x) - 3] => x[0]
print(x[-1])
print(x[-2])
print(x[-3])
````

把序列元素分别导入进变量

````python
point_2d = [10, 20]
x, y = point_2d
print(f'({x}, {y})')

point_3d = [10, 20, 30]
x, y, z = point_3d
print(f'({x}, {y}, {z})')
````

序列中序列元素的获取

````python
x = [1, [2, 3, [4, 5, 6]]]
# 要获取 4
# x[1]          => [2, 3, [4, 5, 6]]
# x[1][2]       => [4, 5, 6]
# x[1][2][0]    => 4
print(x[1][2][0])
````

### 序列中元素的更改
- 通过函数
```
- append(element)       在序列尾端添加值为element元素
- insert(i, element)    在第i位置插入值为element元素
- remove(element)       删除第一个值为element的元素
- pop(i)                删除并且返回第i位置的元素
```
- 通过元素重新定值

````python
a = [1, 2, 3]
a.append(4)
print(a)
a.insert(0, 'hello')
print(a)
a.remove(2)
print(a)
b = a.pop(2)
print(a)
print(b)
a[0] = 'world'      # 把 a 的第一个元素的值变成 'world'
print(a)
````

## 字典
- 以一对数据作为元素的单位, 一般称为键值，利用 : 来分割键和值
- 储存元素的键一定是是独特的，即不会出现两个同样的键
- 不强制数据类型

````python
# a 是拥有三个元素的字典:
# 键为 'john', 值为 100 的键值
# 键为 'Jacky', 值为 0 的键值
# 键为 'malaysia', 值为 singapore 的键值

a = { 'john': 100, 'jacky': 0, 'malaysia': 'singapore' }
````

获取字典拥有指定键元素的值

````python
# person['name'] 就获取 person 中 键 ‘name’ 元素的值 

person = {
    'name': 'john',
    'money': 10,
    'country': 'america',
    'hobbies': ['guitar', 'gaming']
}

money = person['money']
country = person['country']
hobbies = person['hobbies']
````

````python
# 获取是要确保键值是存在的
print(person['gender'])
````

## 字典元素的更改
- 函数
```
- pop(key)  删除并且返回键为key的元素
```
- 元素定值


````python
person = {
    'name': 'fatimah',
    'money': 32,
}     
print(person)
person['name'] = 'aisha'        # 更改名字
print(person)
person['money'] += 10
print(person)
person['gender'] = 'female'     #添加性别的键值
print(person)
money = person.pop('money')
print(person)
print(money)
````
