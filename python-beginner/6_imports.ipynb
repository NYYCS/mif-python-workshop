# CH: 6 Import 用库
- 简单来说，就是用别人写的函数

利用生成随机数的函数作为例子  
- [random库的函数](https://docs.python.org/3/library/random.html)

```python
import random                   # 利用名为 random 的库
print(random.randint(1, 100))   # 生成 1到100的随机数
```

```python
number = 10

count = 0
while True:
    count += 1
    if number == random.randint(1, 100):
        break

print(f'循环了{count}次，才中到{number}')
```

```python
# 从库里面利用指定的函数
from random import randint

print(randint(1, 10))
```

```python
from math import sqrt

def distance(x1, y1, x2, y2):
    delta_x = x2 - x1
    delta_y = y2 - y1
    return sqrt(delta_x * delta_x + delta_y * delta_y)

print(distance(0, 0, 1, 1))
```
