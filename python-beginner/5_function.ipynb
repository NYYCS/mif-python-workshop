# CH 5: 函数
- 把一部分编程过程分隔出去
- 可以接受随意数量的参数
- 可以返回一个值

```python
# 定义一个名为 square 的函数， 它接受一个名为 number 的参数，
# 运行时会返回 number * number

def square(number):
    return number * number 

print(square(3))
```

```python
# 提前定义函数
def say_hello_to(person):
    print(f'Hello {person}!')

people = ['mark', 'steve', 'john']

for person in people:
    say_hello_to(person)
```

```python
# 提供的参数量一定要符合定义函数的参数量
say_hello_to()
```

```python
# 函数可以有默认的参数值
def say_message(message, repeat_for=1):
    for _ in range(repeat_for):
        print(f'we are saying the message: {message}')

say_message('hello world!')
say_message('repeated for 3 times!', 3)
```

提前跳出函数

```python
def give_person_money(person, money):
    alive = person['alive']
    if not alive:
        print('死了还给什么钱')
        return
    name = person['name']
    person['money'] += money
    print(f'{name}拿到了{money}!')

dead_guy = {
    'name': 'john',
    'money': 100000,
    'alive': False
}

give_person_money(dead_guy, 100)
print(dead_guy)
```

### Lambda 函数
这边简单介绍 Lambda 函数
- lambda 函数只是定义函数的另一个写法，
- lambda函数只能运行一行编程

```python

add_one = lambda num: num + 1
# lambda后的num是参数，: 后是运行编程后的返回值

print(add_one(1))

add_two_num = lambda a, b: a + b
# lambda 也可以拥有超过两个参数

print(add_two_num(4, 1))
```
