##### List 序列
```
- 能够存储一序列数据，且不分数据类型
- 是有序的，即数据会依据纳入序列的次序进行排列
```

````python
a = [1, 2, 3]
b = ['a', 'b', 'c']
c = ['hello', 'world', 123, True, [1, 2, 3]]   # 不分数据类型, 不一定要每个元素都同一个数据类型
print(a)
print(b)
print(c)
````

如何知道序列的大小?

````python
print(len(a))
print(len(b))
print(len(c))
````

如何获取列表在第 x 个位置中的值? 
- 元素的位置是从 0 开始算起的
- 获取方式是以 `a_list_variable[x]` 获取的

````python
print(c[0])
print(c[4])
````

````python
print(c[5])       # c 只有五个元素，位置的最大值只能是 len(c) - 1 也就是 4
````

````python
# 负数的位置就是从尾算起的位置
print(c[-3])    # c[-1] => c[len(c) - 1] = c[4]
print(c[-3])    # c[-3] => c[len(c) - 3] = c[2]
````

````python
x = [1, [2, 3, [4, 5, 6]]]

# 要利用位置来获取 5 这个元素呢 ?
# => [1, [2, 3, [4, 5, 6]]]，我们要去 x 的 第 2 元素 => x[1]
# => [2, 3, [4, 5, 6]]， 我们要取 x[1] 的 第 3 个元素  => x[1][2]
# => [4, 5, 6], 我们要取 x[1][2] 的 第 2 个元素 => x[1][2][1]

print(x[1])     
print(x[1][2])  
print(x[1][2][1])
````

如何改变序列?
```
- list_variable[x] = y      # 把在 x 位置的元素变去 y 
- list_variable.append(y)   # 在序列尾端添加值为 y 的元素
- list_variable.pop(x)      # 删除在 x 位置的元素
```

````python
a = [1, 2, 3, 4, 5, 6]
a[0] = 10
print(a)
a.append(100)
print(a)
removed_element = a.pop(0)      # 注意到 a.pop(0) 也等于被删除的元素
print(a)
print(removed_element)
````
