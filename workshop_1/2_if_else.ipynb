 ##### 缩进规则
 一行的 '缩进' 只的是一行根据前头空格的次数所计算的级别  
在 python 里面，只有同样的 '缩进' 的行数在一个中范围才会运行
```

````python
age = 18                    # 缩进数为 0
if age < 18:                
    print('Non-adult')      # 缩进数为 1
    if age < 0:
        print('wtf')
        print('Non-Born')   # 缩进数为 2
elif age < 60:
    print('Adult')          # 缩进数为 1
else:
    print('Senior')         # 缩进数为 1
print('Done')               # 缩进数为 0
````

```
- if        如果
- else      或者
- elif      或者如果 (其实是 else if 的缩写)
```

````python
money = 10
apple_price = 7
have_apple = False
if money - apple_price >= 0:
    # 我够钱买apple
    money -= apple_price
    have_apple = True
    print('我买了apple.')

if have_apple:
    print('我有apple了, yay!')
else:
    print('我没有apple, sad')

print('我现在拥有', money, '块钱.')
````
