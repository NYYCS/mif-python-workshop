#### 变量

````python
# 变量就是保存数据的容器
# x = 10 的意义就是
# 我们设制一个变量，名为 'x‘, 它的值为 10

x = 10
y = 'hello'
print('x is:', x)
print('y is:', y)
````

````python
# 你可以改变变量的值

x = x + 10
y = 'world'
print('x is:', x)
print('y is:', y)
````

````python
x = y
y = 123
print('x is:', x)
print('y is:', y)
````

##### Data Types
变量所包含的数据是有分类型的，主要的类型有:
```
- int    整数
- float  实数
- str    文字 / 字符串
- bool   是/非 
- list   系列     
- dict   字典
```

````python
a = 10                            # int       整数
b = 1.23                          # float     实数
c = 'hello'                       # str       文字 / 字符串 
d = True                          # bool      是/非 
e = [ 10, 20, 'hello', False ]    # list      系列
f = { 'hello': 10 }               # dict      字典  
````

````python
a = '10'        # 这个不是整数！
a += 10         # a 不能加整数10，因为a不是包含整数值的变量
````

##### Casting
换变数据的类型和值

````python
a = int(a)     # int(a) 的意思就是把 a 里面的数据 ('10') 变去整数 (10)
a += 10
print(a)
a = int(2.8)
print(a)       # int(x) 也会把实数换取时取下整数值
````

````python
a = 'hello'   # 当然要换变变量类型时，也要确定变量的数据适合换变
a = int(a)
````

````python
a = 10
a = str(a)  # str(a) 把本来是整数的变量换变去文字
a += 10
````

##### Condition ( Boolean )
```
- x == y    x 等于 y
- x != y    x 不等于 y
- x > y     x 大于 y
- x >= y    x 大于等于 y
- x < y     x 小于 y
- x <= y    x 小于等于 y
```

````python
x = 10
print(x > 1)
print(x == 10)
````

````python
y = 'hello'
z = 'world'
print(x == y)
print(y != z)
z = 'hello'
print(y == z)
````

````python
w = 1 > 2
print(w)
````

Logical Operator
```
- and   和
- or    或
- not   不是
```

````python
print(10 > 5 and 10 > 6)
print(1 > 2 or 1 > 0)
print(not 10 > 5)
x = 100
print((x - 100) == 0 and (x / 5) == 20 or (x * 2) == 100)
````
