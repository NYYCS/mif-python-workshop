假设我们想用程序跟一序列人打招呼  
这是我们只能够写出来的程序

```python
# disclaimer 这是废材的做法！
# 这是凸显循环的重要性的例子罢了

people = ['小明', '小强', '小丽']
print('hello', people[0])
print('hello', people[1])
print('hello', people[2])
```

如果今天 `people` 拥有 100个名字，不可能一个个写出来  
有没有一个更加简单，更加简洁的写法?

##### for - 循环

```python
people = ['小明', '小强', '小丽']
for person in people: 
    print("hello", person) 
```

```python
numbers = [1, 2, 3, 4, 5, 6]

# for number in numbers: 的意思就是
# 对于 numbers 每个元素， 并把元素称呼 number, 
# 注意到 for - loop 也遵守缩进定理

for number in numbers:         
    print(number * number)
    if number > 3:
        print(number, '大于 3')
        
# 循环完 numbers
print('Done!')
```

```python
money = 1000
debts = [100, 200, 350, 1000]

total_debt = 0

for debt in debts:
    total_debt += debt
    if total_debt > money:
        break               # 提前总结循环

if money < total_debt:
    print('不够钱还债')
else:
    print('yay，够钱还')
```

```python
# range(100) 实际上等于 [0, 1, 2, 3, ..., 99] 的序列
for i in range(100):
    if i % 3:
        continue        # 跳过这一次循环
    print(i)
```

假设我们想知道一个数字要被除2几次才会小于1   
用 for-loop 是做不到的 (没有一个非常明确截止循环的重点)

有没有一个写法是可以持续循环，直到我们拿到了我们要的结果?

while - 循环

```python
number = 100
divide_count = 0

# while number > 1: 的意思就是
# 当 number 大于 1 的时候，持续循环

while number > 1:
    number = number / 2
    divide_count += 1

print(divide_count)
```

```python
number = 100
divide_count = 0

# while number > 1: 的意思就是
# 当 number 大于 1 的时候，持续循环

while True:
    if number < 1:
        break           
    number = number / 2
    divide_count += 1

print(divide_count)
```
